```python
students  = []
for line in lines:
    student = line[0:-1].split(",")
    students.append(student)
students[0:5]
```

```python
#3번째 ~ 8번째 컬럼 정수형으로 바꾸고 처음 5줄 출력하기
for student in students:
#     print (student[2] + student[3])
    for i in range(2,8):
        student[i] = int(student[i])
        pass
    pass
    
```
